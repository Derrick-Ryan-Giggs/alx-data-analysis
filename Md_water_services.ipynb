```python
import sqlite3
import re
import os

with open('/home/derrrick-ryan-giggs/Downloads/Additional_download-3901(3)/md_water_services.sql', 'r') as f:
    lines = f.readlines()

cleaned_lines = []
for line in lines:
    # Skip lines we don't need
    if any(x in line for x in ['/*!', 'LOCK TABLES', 'UNLOCK TABLES', 'DROP DATABASE',
                                 'CREATE DATABASE', 'USE `', 'SET @', 'SET NAMES',
                                 'SET TIME_ZONE', 'SET SQL_MODE', 'SET UNIQUE',
                                 'SET FOREIGN', 'DROP TABLE', 'ALTER TABLE',
                                 'DISABLE KEYS', 'ENABLE KEYS']):
        continue

    # Skip MySQL-only constraint lines inside CREATE TABLE
    if re.match(r'\s*KEY\s+', line):
        continue
    if re.match(r'\s*CONSTRAINT\s+', line):
        continue
    if re.match(r'\s*FOREIGN KEY\s+', line):
        continue

    # If previous kept line ended with a comma but next is closing paren, 
    # we may get trailing commas — fix by stripping trailing comma before )
    # Fix escaped quotes \' -> '' (SQLite uses '' to escape single quotes)
    line = line.replace("\\'", "''")

    # Clean end of CREATE TABLE lines: remove ENGINE=... COLLATE=...
    line = re.sub(r'\)\s*ENGINE=\S+.*?;', ');', line)

    # Replace backticks
    line = line.replace('`', '')

    # Remove AUTO_INCREMENT
    line = re.sub(r'\s*AUTO_INCREMENT', '', line)

    cleaned_lines.append(line)

# Fix trailing commas before closing parenthesis in CREATE TABLE
sql = ''.join(cleaned_lines)
sql = re.sub(r',\s*\n\s*\)', '\n)', sql)

# Delete old db
db_path = '/home/derrrick-ryan-giggs/Downloads/Additional_download-3901(3)/md_water_services.db'
if os.path.exists(db_path):
    os.remove(db_path)

conn = sqlite3.connect(db_path)

try:
    conn.executescript(sql)
    conn.commit()
    print(" Done!")
except Exception as e:
    print(f"Error: {e}")
```

```python
import sqlite3
import re
import os

with open('/home/derrrick-ryan-giggs/Downloads/Additional_download-3901(3)/md_water_services.sql', 'r') as f:
    lines = f.readlines()

cleaned_lines = []
for line in lines:
    # Skip MySQL-only admin lines
    if any(x in line for x in [
        '/*!', 'LOCK TABLES', 'UNLOCK TABLES', 'DROP DATABASE',
        'CREATE DATABASE', 'USE `', 'SET @', 'SET NAMES',
        'SET TIME_ZONE', 'SET SQL_MODE', 'SET UNIQUE_CHECKS',
        'SET FOREIGN_KEY_CHECKS', 'DROP TABLE', 'ALTER TABLE',
        'DISABLE KEYS', 'ENABLE KEYS'
    ]):
        continue

    # Skip MySQL-only index/constraint lines inside CREATE TABLE
    if re.match(r'\s*(KEY|UNIQUE KEY|PRIMARY KEY|CONSTRAINT|FOREIGN KEY)\s+`', line):
        continue

    # Fix escaped single quotes MySQL style \' -> SQLite style ''
    line = line.replace("\\'", "''")

    # Strip ENGINE=InnoDB DEFAULT CHARSET=... at end of CREATE TABLE
    line = re.sub(r'\)\s*ENGINE=\S+[^;]*;', ');', line)

    # Replace backticks with nothing
    line = line.replace('`', '')

    # Remove AUTO_INCREMENT
    line = re.sub(r'\s*AUTO_INCREMENT\s*=?\s*\d*', '', line)

    cleaned_lines.append(line)

sql = ''.join(cleaned_lines)

# Fix trailing commas before closing ) in CREATE TABLE blocks
sql = re.sub(r',\s*\n\s*\)', '\n)', sql)

# Rebuild the db fresh
db_path = '/home/derrrick-ryan-giggs/Downloads/Additional_download-3901(3)/md_water_services.db'
if os.path.exists(db_path):
    os.remove(db_path)

conn = sqlite3.connect(db_path)

try:
    conn.executescript(sql)
    conn.commit()
    print("✅ Done! Database loaded.")
except Exception as e:
    print(f"❌ Error: {e}")
```

```python
import sqlite3, pandas as pd

db_path = '/home/derrrick-ryan-giggs/Downloads/Additional_download-3901(3)/md_water_services.db'
conn = sqlite3.connect(db_path)

df = pd.read_sql_query('''
    SELECT 
        ws.type_of_water_source,
        wp.results,
        ROUND(AVG(v.time_in_queue), 1) AS avg_queue_time,
        CASE
            WHEN ws.type_of_water_source = "well" AND wp.results LIKE "%Biological%" 
                THEN "Install UV filter"
            WHEN ws.type_of_water_source = "well" AND wp.results LIKE "%Chemical%" 
                THEN "Install RO filter"
            WHEN ws.type_of_water_source = "shared_tap" AND AVG(v.time_in_queue) > 30 
                THEN "Drill wells"
            WHEN ws.type_of_water_source = "river" 
                THEN "Drill wells"
            WHEN ws.type_of_water_source = "tap_in_home" 
                THEN "Diagnose local infrastructure"
            ELSE "No improvement needed"
        END AS improvement
    FROM water_source ws
    LEFT JOIN well_pollution wp ON ws.source_id = wp.source_id
    JOIN visits v ON ws.source_id = v.source_id
    WHERE v.visit_count = 1
    GROUP BY ws.type_of_water_source, wp.results
    ORDER BY ws.type_of_water_source;
''', conn)

print(df)
conn.close()
```

```python
import sqlite3, pandas as pd

db_path = '/home/derrrick-ryan-giggs/Downloads/Additional_download-3901(3)/md_water_services.db'
conn = sqlite3.connect(db_path)

df = pd.read_sql_query('''
    WITH town_tap_access AS (
        SELECT 
            l.province_name,
            l.town_name,
            ROUND(SUM(CASE WHEN ws.type_of_water_source IN ("tap_in_home", "tap_in_home_broken") 
                      THEN ws.number_of_people_served ELSE 0 END) * 100.0 
                  / SUM(ws.number_of_people_served), 1) AS home_tap_pct
        FROM water_source ws
        JOIN visits v ON ws.source_id = v.source_id
        JOIN location l ON v.location_id = l.location_id
        WHERE v.visit_count = 1
        GROUP BY l.province_name, l.town_name
    )
    SELECT province_name, town_name, home_tap_pct
    FROM town_tap_access
    ORDER BY province_name, home_tap_pct DESC;
''', conn)

print(df)
conn.close()
```
